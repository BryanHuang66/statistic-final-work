```python
import ast
import matplotlib.pyplot as plt
with open('data/abs_GFP.txt','r') as f:
    abs_GFP_string = f.readlines()[0]
```

```python
abs_GFP =list(ast.literal_eval(abs_GFP_string))
```

### Python基本作图 matplotlib

```python
for i in range(len(abs_GFP)):
    plt.plot(*abs_GFP[i])
plt.show()
```

### Python高级作图(Bokeh等库)

```python
with open('data/abs_RFP.txt','r') as f:
    abs_RFP_string = f.readlines()[0]
```

```python
import ast
abs_RFP =list(ast.literal_eval(abs_RFP_string))
```

```python
abs_RFP_list = []
for item in abs_RFP:
    abs_RFP_list.append({"x":item[0],'y':item[1]})
```

```python
## bokeh图片
import pandas as pd
from bokeh.palettes import Spectral5
from bokeh.plotting import figure,show

p = figure(width=800, height=250)
p.title.text = 'ABS RFP PLOT'

for data, name, color in zip(abs_RFP_list, ["first", "second", "third", "forth",'fifth'], Spectral5):
    df = pd.DataFrame(data)
    p.line(df['x'], df['y'], line_width=2, color=color, alpha=0.8, legend_label=name)

p.legend.location = "top_left"
p.legend.click_policy="mute"#hide为完全隐藏，设为mute则是减缓曲线

show(p)
```
